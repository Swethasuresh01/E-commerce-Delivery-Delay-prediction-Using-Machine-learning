### products

```python
path=r"E:\E-commerce_project\dataset ecom\products_dataset.csv"
df_products=pd.read_csv(path)
```

```python
df_products
```

```python
df_products.info()
```

```python
df_products.isnull().sum()
```

```python
# 1. Handle missing values for 'product_category_name' (categorical - impute with mode)
df_products['product_category_name'] = df_products['product_category_name'].fillna(df_products['product_category_name'].mode()[0])
```

```python
df_products.isna().sum()
```

```python
# 2. Handle missing values for 'product_name_length', 'product_description_length', 'product_photos_qty' (numerical - impute with median or mean)
df_products['product_name_lenght'] = df_products['product_name_lenght'].fillna(df_products['product_name_lenght'].median())
df_products['product_description_lenght'] = df_products['product_description_lenght'].fillna(df_products['product_description_lenght'].median())
df_products['product_photos_qty'] = df_products['product_photos_qty'].fillna(df_products['product_photos_qty'].median())
```

```python
df_products.isna().sum()
```

```python
# 3. Handle missing values for 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm' (numerical - impute with mean or median)
df_products['product_weight_g'] = df_products['product_weight_g'].fillna(df_products['product_weight_g'].mean())
df_products['product_length_cm'] = df_products['product_length_cm'].fillna(df_products['product_length_cm'].mean())
df_products['product_height_cm'] = df_products['product_height_cm'].fillna(df_products['product_height_cm'].mean())
df_products['product_width_cm'] = df_products['product_width_cm'].fillna(df_products['product_width_cm'].mean())
```

```python
df_products.isna().sum()
```

```python
df_products.duplicated()
```

```python
# Step 1: Merge orders with customers
df_orders_full1 = pd.merge(df_orders, df_customers, on='customer_id', how='left')
df_orders_full1
```

```python
# Step 2: Merge with order_items to get product_id & seller_id
df_orders_full2 = pd.merge(df_orders_full1, df_items, on='order_id', how='left')
df_orders_full2
```

```python
# Step 3: Merge with products (now we have product_id from order_items)
df_orders_full3 = pd.merge(df_orders_full2, df_products, on='product_id', how='left')
df_orders_full3
```

```python
# Step 4: Merge with product category translation
df_orders_full4= pd.merge(df_orders_full3, df_names, on='product_category_name', how='left')
df_orders_full4
```

```python
# Step 5: Merge with sellers (we have seller_id from order_items)
df_orders_full5 = pd.merge(df_orders_full4, df_sellers, on='seller_id', how='left')
df_orders_full5
```

```python
# Step 6: Merge with payments
df_orders_full6 = pd.merge(df_orders_full5, df_payments, on='order_id', how='left')
df_orders_full6
```

```python
# Step 7: Merge with reviews
df_orders_full7 = pd.merge(df_orders_full6, df_reviews, on='order_id', how='left')
df_orders_full7
```

```python
#convert dataframe to csv
path=r"E:\E-commerce_project\dataset ecom\combined_data.csv"
df_orders_full7.to_csv(path,index=False)
```
