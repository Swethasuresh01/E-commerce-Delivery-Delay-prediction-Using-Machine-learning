```python
# Fill product-related missing values with 0 or suitable placeholders
product_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

df_orders_full7[product_cols] = df_orders_full7[product_cols].fillna(0)
```

```python
# 1. Fill 'unknown' for categorical text columns
df_orders_full7['product_category_name'] = df_orders_full7['product_category_name'].fillna('unknown')
df_orders_full7['product_category_name_english'] = df_orders_full7['product_category_name_english'].fillna('unknown')
df_orders_full7['seller_city'] = df_orders_full7['seller_city'].fillna('unknown')
df_orders_full7['seller_state'] = df_orders_full7['seller_state'].fillna('unknown')
```

```python
# 2. Fill 0 for numerical columns
num_cols = ['price', 'freight_value', 'product_name_lenght', 'product_description_lenght',
            'product_photos_qty', 'product_weight_g', 'product_length_cm',
            'product_height_cm', 'product_width_cm']
df_orders_full7[num_cols] = df_orders_full7[num_cols].fillna(0)
```

```python
# 3. Fill a default string for IDs and other important references (optional but safe)
df_orders_full7['product_id'] = df_orders_full7['product_id'].fillna('missing_prod')
df_orders_full7['seller_id'] = df_orders_full7['seller_id'].fillna('missing_seller')
df_orders_full7['shipping_limit_date'] = df_orders_full7['shipping_limit_date'].fillna('missing_date')
df_orders_full7['seller_zip_code_prefix'] = df_orders_full7['seller_zip_code_prefix'].fillna(-1)
```

```python
# 4. Fix payments
df_orders_full7['payment_sequential'] = df_orders_full7['payment_sequential'].fillna(0).astype(int)
df_orders_full7['payment_type'] = df_orders_full7['payment_type'].fillna('unknown')
df_orders_full7['payment_installments'] = df_orders_full7['payment_installments'].fillna(0).astype(int)
df_orders_full7['payment_value'] = df_orders_full7['payment_value'].fillna(0.0)
```

```python
df_orders_full7 = df_orders_full7[~df_orders_full7['order_item_id'].isnull()]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count top-selling product categories
top_products = df_orders_full7['product_category_name_english'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title('Top 10 Selling Product Categories')
plt.xlabel('Number of Orders')
plt.ylabel('Product Category')
plt.show()
```

```python
# Count top states by number of orders
top_states = df_orders_full7['customer_state'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_states.values, y=top_states.index, palette='magma')
plt.title('Top 10 States by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('State')
plt.show()
```

```python
# Convert the dates to datetime format
df_orders_full7['order_purchase_timestamp'] = pd.to_datetime(df_orders_full7['order_purchase_timestamp'])
df_orders_full7['order_delivered_customer_date'] = pd.to_datetime(df_orders_full7['order_delivered_customer_date'])
df_orders_full7['order_estimated_delivery_date'] = pd.to_datetime(df_orders_full7['order_estimated_delivery_date'])

# Calculate the difference between estimated and actual delivery
df_orders_full7['delivery_delay'] = (df_orders_full7['order_delivered_customer_date'] - df_orders_full7['order_estimated_delivery_date']).dt.days

# Plotting the delivery delays
plt.figure(figsize=(12,6))
sns.histplot(df_orders_full7['delivery_delay'], bins=50, kde=True, color='purple')
plt.title('Delivery Delay Distribution')
plt.xlabel('Days of Delay')
plt.ylabel('Frequency')
plt.show()
```

```python
# 1. Average Delay
average_delay = df_orders_full7['delivery_delay'].mean()
print(f"Average Delivery Delay: {average_delay} days")

# 2. Percentage of On-Time Deliveries (where delivery_delay <= 0)
on_time_percentage = (df_orders_full7[df_orders_full7['delivery_delay'] <= 0].shape[0] / df_orders_full7.shape[0]) * 100
print(f"Percentage of On-Time Deliveries: {on_time_percentage:.2f}%")

# 3. Orders with High Delays (for example, greater than 7 days)
high_delay_orders = df_orders_full7[df_orders_full7['delivery_delay'] > 7]
print(f"Orders with delays greater than 7 days: {high_delay_orders.shape[0]}")
```

```python
# Calculate delay by state
df_orders_full7['delivery_delay'] = (df_orders_full7['order_delivered_customer_date'] - df_orders_full7['order_estimated_delivery_date']).dt.days

# Group by state and calculate the average delay
state_delay = df_orders_full7.groupby('customer_state')['delivery_delay'].mean().sort_values(ascending=False)

# Plot the delay by state
plt.figure(figsize=(12,6))
state_delay.plot(kind='bar', color='orange')
plt.title('Average Delivery Delay by State')
plt.xlabel('State')
plt.ylabel('Average Delivery Delay (days)')
plt.show()
```

```python
# Group by product category and calculate the average delay
category_delay = df_orders_full7.groupby('product_category_name')['delivery_delay'].mean().sort_values(ascending=False)

# Plot the delay by product category
plt.figure(figsize=(12,6))
category_delay.plot(kind='bar', color='green')
plt.title('Average Delivery Delay by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Delivery Delay (days)')
plt.show()
```

```python
# Group by payment type and calculate the average delivery delay
payment_delay = df_orders_full7.groupby('payment_type')['delivery_delay'].mean().sort_values(ascending=False)

# Plot the delay by payment type
plt.figure(figsize=(12,6))
payment_delay.plot(kind='bar', color='blue')
plt.title('Average Delivery Delay by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Delivery Delay (days)')
plt.show()
```

```python
# Group by order status and calculate the average delivery delay
status_delay = df_orders_full7.groupby('order_status')['delivery_delay'].mean().sort_values(ascending=False)

# Plot the delay by order status
plt.figure(figsize=(12,6))
status_delay.plot(kind='bar', color='purple')
plt.title('Average Delivery Delay by Order Status')
plt.xlabel('Order Status')
plt.ylabel('Average Delivery Delay (days)')
plt.show()
```

```python
# Correlation matrix for numerical variables
numerical_cols = df_orders_full7.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df_orders_full7[numerical_cols].corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# Calculate total order value per customer
customer_order_value = df_orders_full7.groupby('customer_id')['price'].sum().reset_index()

# Segment customers into three categories: Low, Medium, High
bins = [0, 100, 500, float('inf')]
labels = ['Low', 'Medium', 'High']
customer_order_value['segment'] = pd.cut(customer_order_value['price'], bins=bins, labels=labels)

# View the segmentation
print(customer_order_value['segment'].value_counts())
```

```python
# Grouping by product category and calculating average price
product_price_analysis = df_orders_full7.groupby('product_category_name')['price'].mean().reset_index()

# Sorting by price to find the top and low-priced categories
product_price_analysis_sorted = product_price_analysis.sort_values(by='price', ascending=False)

# Visualizing the top 10 most expensive product categories
plt.figure(figsize=(10,6))
sns.barplot(x='price', y='product_category_name', data=product_price_analysis_sorted.head(10))
plt.title("Top 10 Expensive Product Categories")
plt.show()
```

```python
# Sales Metrics
total_sales = df_orders_full7['payment_value'].sum()
average_order_value = df_orders_full7['payment_value'].mean()

# Customer Metrics
repeat_customers = df_orders_full7['customer_unique_id'].value_counts()
repeat_customer_rate = (repeat_customers[repeat_customers > 1].count() / df_orders_full7['customer_unique_id'].nunique()) * 100

# Delivery Metrics
on_time_deliveries = df_orders_full7[df_orders_full7['delivery_delay'] <= 0].shape[0]
on_time_delivery_rate = (on_time_deliveries / df_orders_full7.shape[0]) * 100
average_delivery_delay = df_orders_full7['delivery_delay'].mean()

# Product Metrics
top_selling_products = df_orders_full7.groupby('product_category_name_english')['payment_value'].sum().sort_values(ascending=False).head(10)
returned_orders = df_orders_full7[df_orders_full7['order_status'] == 'returned'].shape[0]
product_return_rate = (returned_orders / df_orders_full7.shape[0]) * 100

# Payment Metrics
payment_method_distribution = df_orders_full7['payment_type'].value_counts(normalize=True) * 100
successful_orders = df_orders_full7[df_orders_full7['order_status'].isin(['delivered', 'shipped'])].shape[0]
payment_success_rate = (successful_orders / df_orders_full7.shape[0]) * 100

# Print Metrics
print("Total Sales:", round(total_sales, 2))
print("Average Order Value:", round(average_order_value, 2))
print("Repeat Customer Rate:", round(repeat_customer_rate, 2), "%")
print("On-Time Delivery Rate:", round(on_time_delivery_rate, 2), "%")
print("Average Delivery Delay:", round(average_delivery_delay, 2), "days")
print("Top-Selling Products:\n", top_selling_products)
print("Product Return Rate:", round(product_return_rate, 2), "%")
print("Payment Method Distribution:\n", payment_method_distribution)
print("Payment Success Rate:", round(payment_success_rate, 2), "%")
```
